### 特徵工程
#### 數據前處理
    1. 無量綱化
    2. 對定量特徵二值化
    3. 對定性特徵啞編碼
    4. 缺失值計算
    5. 數據變換
#### 特徵選擇
    1. Filter
    2. Wrapper
    3. Embedded
#### 降維
    1. 主成分分析法（PCA）
    2. 線性判別分析法（LDA)

```python
#取得資料集
from sklearn.datasets import load_iris

iris = load_iris()

#特徵矩陣
print(iris.data)

#目標向量
print(iris.target)
```

#### 數據前處理

```python
#數據前處理
#標準化
from sklearn.preprocessing import StandardScaler

StandardScaler().fit_transform(iris.data)
```

```python
#區間縮放 設定最大最小值 縮放到此數值區間中
from sklearn.preprocessing import MinMaxScaler

MinMaxScaler().fit_transform(iris.data)
```

```python
#一般化or正規化
from sklearn.preprocessing import Normalizer

Normalizer().fit_transform(iris.data)
```

```python
#對定性特徵進行One-hot encoding
from sklearn.preprocessing import OneHotEncoder

#對目標欄位資料進行One-hot encoding
OneHotEncoder().fit_transform(iris.target.reshape((-1,1)))
```

```python
#缺失值計算
from numpy import vstack, array, nan
from sklearn.impute import SimpleImputer

#參數missing_value 為缺失值的表示形式 預設為NaN
#參數strategy 為缺失值的補值方式 預設為Mean
SimpleImputer().fit_transform(vstack((array([nan, nan, nan, nan]),iris.data)))
```

#### 特徵選擇

```python
#Filter
#變異數選擇法
from sklearn.feature_selection import VarianceThreshold

#參數threshold可以設定方差之閾值
VarianceThreshold(threshold=3).fit_transform(iris.data)
```

```python
#卡方檢驗 檢驗定性自變量對因變量的相關性 用於處理類別型資料
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#選擇最好的k個特徵
SelectKBest(chi2, k=2).fit_transform(iris.data, iris.target)
```

```python
#相互資訊法
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import mutual_info_classif

#SelectKBest(lambda X, Y: array(list(map(lambda x:mic(x,Y), X.T))).T, k=2).fit_transform(iris.data, iris.target)
SelectKBest(mutual_info_classif, k=2).fit_transform(iris.data, iris.target)
```

```python
#Wrapper 選擇一個基本模型進行多輪訓練 選出最佳特徵子集
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

#參數estimator為基本模型
#參數n_features_to_select為選擇的特徵個數
RFE(estimator=LogisticRegression(), n_features_to_select=2).fit_transform(iris.data, iris.target)
```

```python
#Embedded 帶有懲罰項 藉此自動進行特徵選擇
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression

#參數estimator為基本模型
#參數n_features_to_select為選擇的特徵個數
SelectFromModel(LogisticRegression(penalty="l2", C=0.1)).fit_transform(iris.data, iris.target)
```

#### 降維

```python
#PCA 
from sklearn.decomposition import PCA

#參數n_components為主成份數目
PCA(n_components=2).fit_transform(iris.data)
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

#參數n_components為降維後的維數
LDA(n_components=2).fit_transform(iris.data, iris.target)
```
